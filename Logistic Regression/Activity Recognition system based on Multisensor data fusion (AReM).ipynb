# create function to Read & Concatenate CSVs file

```python
def read_and_concat_csvs(folder, new_col_name, value):
    dfs = []
    for file in os.listdir(folder):
        if file.endswith('.csv'):
            path = os.path.join(folder, file)
            
            with open(path, 'r') as f:
                lines = f.readlines()

            # Extract columns from '# Columns:' line
            columns = None
            for line in lines:
                if line.startswith('# Columns:'):
                    # Assuming columns are separated by commas or tabs, adjust as needed
                    columns = line.strip().replace('# Columns:', '').strip().split(',')
                    
                    break

            # Read the CSV, skipping 5 rows as in your example
            df = pd.read_csv(path, delimiter=',', skiprows=5, on_bad_lines='skip', names=columns)
            df[new_col_name] = value
            dfs.append(df)
    
    combined_df = pd.concat(dfs, ignore_index=True)
    return combined_df
```

```python
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
```

```python
def plot_multiclass_roc(clf, x_test, y_test, n_classes, figsize=(5,5)):
    y_score = clf.predict_proba(x_test)

    # structures
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # calculate dummies once
    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # roc for each class
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('Receiver operating characteristic example')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="lower right",fontsize=8)
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
```

```python
plot_multiclass_roc(lr, x_scaled_test, y_test, n_classes = 7)
```
